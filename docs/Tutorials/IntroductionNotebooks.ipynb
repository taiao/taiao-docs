## NumPy

*NumPy* is a library supporting large, multi-dimensional arrays and matrices, providing high-level mathematical functions. 


* *ndarray*: a homogeneous n-dimensional array object

![NumPy](img/numpy.png)

## NumPy Example

```python
import numpy as np

a = np.array([1, 2, 3])   # Create a rank 1 array"
print(np.sum(a))          # Prints 1+2+3
```

```python
print(a.shape)            # Prints "(3,)"
print(a[0], a[1], a[2])   # Prints "1 2 3"
```

```python
a[0] = 5                  # Change an element of the array
print(a)                  # Prints "[5, 2, 3]"
```

```python
b = np.array([[1,2,3],[4,5,6]])    # Create a rank 2 array
print(b.shape)                     # Prints "(2, 3)"
print(b[0, 0], b[0, 1], b[1, 0])   # Prints "1 2 4"
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
g = sns.pairplot(data.drop('Id', axis=1), hue='Species', markers='+')
plt.show()
```

```python
g = sns.violinplot(y='Species', x='SepalLengthCm', data=data, inner='quartile')
plt.show()
```
